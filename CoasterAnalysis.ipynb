###Question 1  

What are the location with the fastest roller coasters (Minimum of 10 roller coasters)

```python
df.head()
```

```python
df['Location'].value_counts()
```

```python
ax = df.query("Location != 'Other'") \
.groupby(['Location'])['Speed_MPH'] \
.agg(['mean', 'count']) \
.query( 'count >= 10') \
.sort_values('mean', ascending=True) ['mean'] \
.plot(kind='barh', title='Average speed by Location')

ax.set_xlabel('Average Speed');
```

### Question 2

Which decade show the highest growth rate in roller coasters industry.

```python
df.head()
```

```python
df['Year_Introduced'].unique()
```

```python
bins = [1880,1890,1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020,2030]
labels = ['1880s','1890s','1900s','1910s','1920s','1930s','1940s','1950s','1960s','1970s','1980s','1990s','2000s','2010s','2020s']
```

```python
df['Decade'] = pd.cut(df['Year_Introduced'], bins=bins, labels=labels)
df
```

```python
ax = df['Decade'].value_counts().plot(kind='bar', title='Maximum number of inauguration of Rollercoaster in each decade');
ax.set_xlabel('Decade')
ax.set_ylabel('Count')
ax.set_xticklabels(labels);
```

```python
abc = df['Decade'].value_counts()
round((((abc [1] / abc[0]) - 1) * 100), 2)
```
